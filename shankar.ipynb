```python
import numpy as np
import pandas as pd 
from seaborn import countplot
from matplotlib.pyplot import figure
from sklearn import metrics,linear_model
#from sklearn import databases
from sklearn.neighbors import KNeighborsClassifier
```

```python
Symptums = pd.read_csv('data_pivoted.csv')
print(Symptums.head())

```

```python
Symptums
```

```python
x=Symptums.drop("disease",axis=1)

y=Symptums["disease"]
```

```python
x
```

```python
y
```

```python
print("--------------trainning data of Symptums--------------")
print(x.head())
print("--------------trainning data of Symptums--------------")
print(y.head())
x=x.fillna(0)
y=y.fillna(0)
print(x)
```

```python
x_train=x
y_train=y

x_test=x
y_test=y

knn = KNeighborsClassifier(n_neighbors=1)

knn.fit(x_train, y_train)

print("trainning complete")
```

```python
print(x_train)
```

```python
prediction=knn.predict(x_test)

print(prediction)
```

```python
print("Accuracy",metrics.accuracy_score(y_test,prediction))
```
